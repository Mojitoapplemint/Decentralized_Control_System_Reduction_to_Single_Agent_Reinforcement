```python
import numpy as np
import pandas as pd
import gymnasium as gym
import cam_env_5050_entry
import cam_env_cat_entry

MOUSE_STATES = {
        (1,3):0,
        (1,4):1, (1,5):1,
        (2,3):2,
        (2,4):3, (2,5):3,
        (3,3):4,
        (3,4):5, (3,5):5
    }

MOUSE_DOORS = ["m1", "m2", "m3", "c2", "c3"]

CAT_STATES = {
        (1,3):0, (2,3):0,
        (1,4):1, (2,4):1, 
        (1,5):2, (2,5):2, 
        (3,3):3,
        (3,4):4, 
        (3,5):5
    }

CAT_DOORS = ["m2", "m3", "c1", "c2", "c3"]
```

```python
def get_cat_state_num(observation, door_status):
    return 4*CAT_STATES.get(observation)+door_status


def get_mouse_state_num(observation, door_status):
    return 4*MOUSE_STATES.get(observation)+door_status


def binary_tuple_to_decimal(door_status):
    return door_status[0]*2 + door_status[1]


def doors_mouse_projection(mouse_door, cat_door):
    if cat_door[0] in MOUSE_DOORS:
        return (mouse_door[1], cat_door[1])
    else:
        return (mouse_door[1], 1)

    
def doors_cat_projection(mouse_door, cat_door):
    if mouse_door[0] in CAT_DOORS:
        return (mouse_door[1], cat_door[1])
    else:
        return (1, cat_door[1])
```

```python
q_cat = pd.read_csv("iql_5050_entry_cat.csv")
q_cat = q_cat.drop(q_cat.columns[[0]], axis=1).to_numpy()

q_mouse = pd.read_csv("iql_5050_entry_mouse.csv")
q_mouse = q_mouse.drop(q_mouse.columns[[0]], axis=1).to_numpy()

env = gym.make("CatAndMouse-5050_entry", render_mode = "human")

terminated = False
truncated = False

observation, info = env.reset()

while(not terminated and not truncated):
    mouse_door = info["mouse_doors"]
    cat_door = info["cat_doors"]
    
    mouse_door_observation = doors_mouse_projection(mouse_door, cat_door)
    cat_door_observation = doors_cat_projection(mouse_door, cat_door)
    
    mouse_door_status = binary_tuple_to_decimal(mouse_door_observation)
    cat_door_status = binary_tuple_to_decimal(cat_door_observation)
    
    mouse_state_num = get_mouse_state_num(observation, mouse_door_status)
    cat_state_num = get_cat_state_num(observation, cat_door_status)
    
    cat_action = np.argmax(q_cat[cat_state_num])
    mouse_action = np.argmax(q_mouse[mouse_state_num])
    
    joint_action = (mouse_action, cat_action)
    
    observation, _, terminated, truncated, info = env.step(joint_action)
    
```

```python
q_cat = pd.read_csv("iql_cat_entry_cat.csv")
q_cat = q_cat.drop(q_cat.columns[[0]], axis=1).to_numpy()

q_mouse = pd.read_csv("iql_cat_entry_mouse.csv")
q_mouse = q_mouse.drop(q_mouse.columns[[0]], axis=1).to_numpy()

env = gym.make("CatAndMouse-cat_entry", render_mode = "human")

terminated = False
truncated = False

observation, info = env.reset()

while(not terminated and not truncated):
    cat_state_num = get_cat_state_num(observation)
    mouse_state_num = get_mouse_state_num(observation)
    
    cat_action = np.argmax(q_cat[cat_state_num])
    mouse_action = np.argmax(q_mouse[mouse_state_num])
    
    joint_action = (mouse_action, cat_action)
    
    observation, _, terminated, truncated, info = env.step(joint_action)
```
